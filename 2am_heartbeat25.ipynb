```python
#======================================================
# Full merged notebook script (Colab-friendly)
# ======================================================
# Install required libraries (run once in Colab)
# (Some packages are optional; install if you want those features)
!pip install -q imbalanced-learn shap lightgbm xgboost torch torchvision PyALE scikit-learn>=1.0

# ------------------------------
# 1) Setup & imports
# ------------------------------
import os, random, warnings
warnings.filterwarnings("ignore")
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")

from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.metrics import (
    roc_auc_score, average_precision_score, accuracy_score,
    precision_score, recall_score, f1_score, roc_curve,
    precision_recall_curve, confusion_matrix, # calibration_curve, # Removed due to import error
    brier_score_loss
)
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, VotingClassifier, StackingClassifier
from sklearn.tree import DecisionTreeClassifier, export_text
from sklearn.inspection import PartialDependenceDisplay

import xgboost as xgb
import lightgbm as lgb
from imblearn.over_sampling import SMOTE

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, BatchNormalization, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping

import torch, torch.nn as nn
from torch.utils.data import Dataset, DataLoader

import shap
import joblib

# Optional imports (may need pip install)
_DELONG_AVAILABLE = False # delong is causing issues, skip for now
# try:
#     from delong import delong_roc_test # Removed due to import error
#     _DELONG_AVAILABLE = True
# except Exception:
#     _DELONG_AVAILABLE = False

try:
    from PyALE import ale
    _PYALE_AVAILABLE = True
except Exception:
    _PYALE_AVAILABLE = False

# 1b) Reproducibility & outdir
# ------------------------------
SEED = 42
random.seed(SEED)
np.random.seed(SEED)
os.environ["PYTHONHASHSEED"] = str(SEED)
tf.random.set_seed(SEED)
torch.manual_seed(SEED)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(SEED)

OUTDIR = "outputs"
os.makedirs(OUTDIR, exist_ok=True)

print("Environment ready. DELONG:", _DELONG_AVAILABLE, "PyALE:", _PYALE_AVAILABLE)

# ------------------------------
# 2) Helper functions
# ------------------------------
results_store = {}

def predict_proba_safe(model_or_probs, X):
    """Return probability array for positive class.
    Accepts sklearn-like model (predict_proba), Keras model (predict), PyTorch model should have produced probs already,
    or a raw numpy array of probabilities."""
    # raw probs array
    if isinstance(model_or_probs, (list, np.ndarray)):
        return np.array(model_or_probs).astype(float).flatten()
    model = model_or_probs
    if hasattr(model, "predict_proba"):
        return model.predict_proba(X)[:, 1]
    if hasattr(model, "predict"):
        preds = model.predict(X)
        return np.array(preds).flatten()
    raise ValueError("Object provided has no predict/predict_proba and is not an array.")

def threshold_analysis(probs, y, ths=[0.3,0.4,0.5,0.6]):
    out = {}
    probs = np.array(probs).astype(float).flatten()
    for t in ths:
        preds = (probs >= t).astype(int)
        out[float(t)] = {
            "acc": float(accuracy_score(y, preds)),
            "prec": float(precision_score(y, preds, zero_division=0)),
            "rec": float(recall_score(y, preds, zero_division=0)),
            "f1": float(f1_score(y, preds, zero_division=0)),
            "auc": float(roc_auc_score(y, probs)),
            "ap": float(average_precision_score(y, probs))
        }
    prec, rec, thr = precision_recall_curve(y, probs)
    f1s = 2 * (prec * rec) / (prec + rec + 1e-12)
    best_idx = int(np.nanargmax(f1s))
    best_thresh = float(thr[best_idx]) if best_idx < len(thr) else 0.5
    out["best_f1_thresh"] = best_thresh
    out["best_f1"] = float(np.nanmax(f1s))
    return out

def eval_and_store(name, probs_val, probs_test, y_val, y_test):
    tv = threshold_analysis(probs_val, y_val)
    tt = threshold_analysis(probs_test, y_test)
    res = {
        "val": tv,
        "test": tt,
        "auc_val": float(roc_auc_score(y_val, probs_val)),
        "auc_test": float(roc_auc_score(y_test, probs_test)),
        "ap_val": float(average_precision_score(y_val, probs_val)),
        "ap_test": float(average_precision_score(y_test, probs_test)),
    }
    results_store[name] = res
    print(f"[EVAL] {name} â€” AUC val/test: {res['auc_val']:.4f}/{res['auc_test']:.4f}")

def plot_dashboard(name, probs_val, probs_test, y_val, y_test, thresh=0.5):
    fig, axes = plt.subplots(2,2, figsize=(12,8))
    ax_roc, ax_pr, ax_cm, ax_cal = axes.flatten()

    # ROC
    fpr, tpr, _ = roc_curve(y_test, probs_test)
    ax_roc.plot(fpr, tpr, label=f"AUC={roc_auc_score(y_test, probs_test):.3f}")
    ax_roc.plot([0,1],[0,1], 'k--', linewidth=0.7)
    ax_roc.set_title("ROC"); ax_roc.set_xlabel("FPR"); ax_roc.set_ylabel("TPR"); ax_roc.legend()

    # PR
    prec, rec, _ = precision_recall_curve(y_test, probs_test)
    ax_pr.plot(rec, prec, label=f"AP={average_precision_score(y_test, probs_test):.3f}")
    ax_pr.set_title("Precision-Recall"); ax_pr.set_xlabel("Recall"); ax_pr.set_ylabel("Precision"); ax_pr.legend()

    # Confusion matrix
    preds = (np.array(probs_test) >= thresh).astype(int)
    cm = confusion_matrix(y_test, preds)
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", ax=ax_cm)
    ax_cm.set_title(f"Confusion Matrix @th={thresh:.2f}")
    ax_cm.set_xlabel("Predicted"); ax_cm.set_ylabel("True")

    # Calibration
    # frac_pos, mean_pred = calibration_curve(y_test, probs_test, n_bins=10) # Removed due to import error
    # ax_cal.plot(mean_pred, frac_pos, marker='o', label=name) # Removed due to import error
    ax_cal.plot([0,1],[0,1], 'k--', linewidth=0.7) # Keep ideal line
    ax_cal.set_title("Calibration (Plotting Skipped)"); ax_cal.set_xlabel("Mean pred prob"); ax_cal.set_ylabel("Fraction positive")
    # ax_cal.legend() # Removed as no models are plotted

    plt.suptitle(f"Model: {name} (threshold={thresh:.2f})")
    plt.tight_layout(rect=[0,0,1,0.96])
    plt.show()

def expected_calibration_error(y_true, y_prob, n_bins=10):
    y_prob = np.array(y_prob).flatten()
    bins = np.linspace(0.0, 1.0, n_bins+1)
    ece = 0.0
    for i in range(n_bins):
        idx = (y_prob >= bins[i]) & (y_prob < bins[i+1])
        if np.any(idx):
            acc_bin = float(np.mean(y_true[idx]))
            conf_bin = float(np.mean(y_prob[idx]))
            ece += np.abs(acc_bin - conf_bin) * (idx.sum() / len(y_true))
    return float(ece)

# ------------------------------
# 3) Data load / preprocessing
# ------------------------------
# Adjust this path if your CSV is elsewhere
DATA_PATH = "/content/heart_disease_health_indicators_BRFSS2015.csv"
if not os.path.exists(DATA_PATH):
    raise FileNotFoundError(f"Dataset not found at {DATA_PATH}; upload it to the working directory or change DATA_PATH.")

df = pd.read_csv(DATA_PATH)
# Feature lists (same as your repo)
binary_cols = ["HighBP","HighChol","CholCheck","Smoker","Stroke","Diabetes",
               "PhysActivity","Fruits","Veggies","HvyAlcoholConsump",
               "AnyHealthcare","NoDocbcCost","DiffWalk","Sex"]
numeric_cols = ["BMI","MentHlth","PhysHlth","Age"]
categorical_cols = ["GenHlth","Education","Income"]
FEAT_NAMES = binary_cols + numeric_cols + categorical_cols
TARGET = "HeartDiseaseorAttack"

# Basic splits
X_raw = df[FEAT_NAMES].copy()
y_raw = df[TARGET].astype(int).copy()

X_train_raw, X_temp_raw, y_train_raw, y_temp_raw = train_test_split(
    X_raw, y_raw, test_size=0.30, stratify=y_raw, random_state=SEED
)
X_val_raw, X_test_raw, y_val_raw, y_test_raw = train_test_split(
    X_temp_raw, y_temp_raw, test_size=0.50, stratify=y_temp_raw, random_state=SEED
)

# Preprocessing: numeric impute+scale, categorical impute (keeps binary as-is)
num_transformer = Pipeline([("imputer", SimpleImputer(strategy="median")), ("scaler", StandardScaler())])
cat_transformer = Pipeline([("imputer", SimpleImputer(strategy="most_frequent"))])
preprocessor = ColumnTransformer(transformers=[
    ("num", num_transformer, numeric_cols),
    ("cat", cat_transformer, binary_cols + categorical_cols)
], remainder='drop')

preprocessor.fit(X_train_raw)
X_train = preprocessor.transform(X_train_raw)
X_val = preprocessor.transform(X_val_raw)
X_test = preprocessor.transform(X_test_raw)
# convert if sparse
try:
    if hasattr(X_train, "toarray"):
        X_train = X_train.toarray(); X_val = X_val.toarray(); X_test = X_test.toarray()
except Exception:
    pass

# Balance training set with SMOTE
sm = SMOTE(random_state=SEED)
X_train_res, y_train_res = sm.fit_resample(X_train, y_train_raw)
print("Done preprocessing. Shapes:", X_train_res.shape, X_val.shape, X_test.shape)

# ------------------------------
# 4) Classical ML + DL + TabTransformer + Ensemble
#    â€” each section trains model, computes probs, eval_and_store, and plot dashboard
# ------------------------------

# 4.1 Logistic Regression
logreg = LogisticRegression(max_iter=1000, class_weight="balanced", random_state=SEED)
logreg.fit(X_train_res, y_train_res)
probs_log_val = predict_proba_safe(logreg, X_val)
probs_log_test = predict_proba_safe(logreg, X_test)
eval_and_store("LogReg", probs_log_val, probs_log_test, y_val_raw, y_test_raw)
plot_dashboard("LogReg", probs_log_val, probs_log_test, y_val_raw, y_test_raw, thresh=0.5)

# 4.2 Random Forest
rf = RandomForestClassifier(n_estimators=300, class_weight="balanced", random_state=SEED, n_jobs=-1)
rf.fit(X_train_res, y_train_res)
probs_rf_val = predict_proba_safe(rf, X_val)
probs_rf_test = predict_proba_safe(rf, X_test)
eval_and_store("RandomForest", probs_rf_val, probs_rf_test, y_val_raw, y_test_raw)
plot_dashboard("RandomForest", probs_rf_val, probs_rf_test, y_val_raw, y_test_raw, thresh=0.5)

# 4.3 XGBoost
xgb_model = xgb.XGBClassifier(use_label_encoder=False, eval_metric="logloss", random_state=SEED, n_jobs=-1)
xgb_model.fit(X_train_res, y_train_res, eval_set=[(X_val, y_val_raw)], verbose=False)
probs_xgb_val = predict_proba_safe(xgb_model, X_val)
probs_xgb_test = predict_proba_safe(xgb_model, X_test)
eval_and_store("XGBoost", probs_xgb_val, probs_xgb_test, y_val_raw, y_test_raw)
plot_dashboard("XGBoost", probs_xgb_val, probs_xgb_test, y_val_raw, y_test_raw, thresh=0.5)

# 4.4 LightGBM
lgb_model = lgb.LGBMClassifier(random_state=SEED, n_jobs=-1)
lgb_model.fit(X_train_res, y_train_res, eval_set=[(X_val, y_val_raw)])
probs_lgb_val = predict_proba_safe(lgb_model, X_val)
probs_lgb_test = predict_proba_safe(lgb_model, X_test)
eval_and_store("LightGBM", probs_lgb_val, probs_lgb_test, y_val_raw, y_test_raw)
plot_dashboard("LightGBM", probs_lgb_val, probs_lgb_test, y_val_raw, y_test_raw, thresh=0.5)

# 4.5 ANN (simple)
input_dim = X_train_res.shape[1]
ann = Sequential([Dense(32, activation="relu", input_shape=(input_dim,)), Dense(1, activation="sigmoid")])
ann.compile(optimizer=Adam(1e-3), loss="binary_crossentropy")
ann.fit(X_train_res, y_train_res, validation_data=(X_val, y_val_raw), epochs=20, batch_size=512,
        callbacks=[EarlyStopping(patience=5, restore_best_weights=True)], verbose=1)
probs_ann_val = ann.predict(X_val).flatten()
probs_ann_test = ann.predict(X_test).flatten()
eval_and_store("ANN", probs_ann_val, probs_ann_test, y_val_raw, y_test_raw)
plot_dashboard("ANN", probs_ann_val, probs_ann_test, y_val_raw, y_test_raw, thresh=0.5)

# 4.6 DeepMLP (deeper)
def get_deep_mlp(input_dim):
    model = Sequential([
        Dense(256, activation="relu", input_shape=(input_dim,)), BatchNormalization(), Dropout(0.3),
        Dense(128, activation="relu"), BatchNormalization(), Dropout(0.3),
        Dense(64, activation="relu"), BatchNormalization(), Dropout(0.2),
        Dense(32, activation="relu"),
        Dense(1, activation="sigmoid")
    ])
    model.compile(optimizer=Adam(1e-3), loss="binary_crossentropy")
    return model

deep_mlp = get_deep_mlp(input_dim)
deep_mlp.fit(X_train_res, y_train_res, validation_data=(X_val, y_val_raw),
             epochs=50, batch_size=512, callbacks=[EarlyStopping(patience=7, restore_best_weights=True)], verbose=1)
probs_dmlp_val = deep_mlp.predict(X_val).flatten()
probs_dmlp_test = deep_mlp.predict(X_test).flatten()
eval_and_store("DeepMLP", probs_dmlp_val, probs_dmlp_test, y_val_raw, y_test_raw)
plot_dashboard("DeepMLP", probs_dmlp_val, probs_dmlp_test, y_val_raw, y_test_raw, thresh=0.5)
```

```python

# 4.7 TabTransformer (simple PyTorch implementation)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

class TabDatasetTorch(Dataset):
    def __init__(self, X_np, y_np):
        self.X = torch.tensor(X_np, dtype=torch.float32)
        self.y = torch.tensor(y_np.values if hasattr(y_np, "values") else y_np, dtype=torch.float32).unsqueeze(1)
    def __len__(self): return len(self.y)
    def __getitem__(self, idx): return self.X[idx], self.y[idx]

class SimpleTabTransformer(nn.Module):
    def __init__(self, input_dim, emb_dim=32, n_heads=4, n_layers=2, mlp_hidden=128, dropout=0.1):
        super().__init__()
        self.proj = nn.Linear(1, emb_dim)
        enc_layer = nn.TransformerEncoderLayer(d_model=emb_dim, nhead=n_heads, dropout=dropout)
        self.transformer = nn.TransformerEncoder(enc_layer, num_layers=n_layers)
        self.mlp = nn.Sequential(
            nn.Linear(input_dim * emb_dim, mlp_hidden),
            nn.ReLU(),
            nn.Dropout(dropout),
            nn.Linear(mlp_hidden, 1),
            nn.Sigmoid()
        )
    def forward(self, x):
        B, D = x.shape
        x_proj = self.proj(x.unsqueeze(-1))          # (B, D, E)
        x_t = x_proj.permute(1,0,2)                 # (D, B, E)
        out = self.transformer(x_t)                 # (D, B, E)
        out = out.permute(1,0,2).reshape(B, -1)     # (B, D*E)
        return self.mlp(out)

tab_model = SimpleTabTransformer(X_train_res.shape[1]).to(device)
optimizer = torch.optim.Adam(tab_model.parameters(), lr=1e-3)
loss_fn = nn.BCELoss()
train_ds = TabDatasetTorch(X_train_res, y_train_res)
train_loader = DataLoader(train_ds, batch_size=512, shuffle=True)
tab_model.train()
for epoch in range(5):
    running = 0.0
    for xb, yb in train_loader:
        xb, yb = xb.to(device), yb.to(device)
        optimizer.zero_grad()
        preds = tab_model(xb)
        loss = loss_fn(preds, yb)
        loss.backward()
        optimizer.step()
        running += loss.item()
    print(f"[TabTransformer] Epoch {epoch} loss: {running/len(train_loader):.4f}")

tab_model.eval()
with torch.no_grad():
    probs_tab_val = tab_model(torch.tensor(X_val, dtype=torch.float32).to(device)).cpu().numpy().flatten()
    probs_tab_test = tab_model(torch.tensor(X_test, dtype=torch.float32).to(device)).cpu().numpy().flatten()
eval_and_store("TabTransformer", probs_tab_val, probs_tab_test, y_val_raw, y_test_raw)
plot_dashboard("TabTransformer", probs_tab_val, probs_tab_test, y_val_raw, y_test_raw, thresh=0.5)

# 4.8 Voting Ensemble (soft)
ensemble = VotingClassifier(
    estimators=[
        ("xgb", xgb.XGBClassifier(use_label_encoder=False, eval_metric="logloss", random_state=SEED)),
        ("rf", RandomForestClassifier(n_estimators=300, random_state=SEED, class_weight="balanced")),
        ("lgb", lgb.LGBMClassifier(random_state=SEED))
    ],
    voting="soft", n_jobs=-1
)
ensemble.fit(X_train_res, y_train_res)
probs_ens_val = predict_proba_safe(ensemble, X_val)
probs_ens_test = predict_proba_safe(ensemble, X_test)
eval_and_store("Ensemble", probs_ens_val, probs_ens_test, y_val_raw, y_test_raw)
plot_dashboard("Ensemble", probs_ens_val, probs_ens_test, y_val_raw, y_test_raw, thresh=0.5)

# ------------------------------
# 5) Combined Visualizations and Results table
# ------------------------------
models_probs = {
    "LogReg": probs_log_test,
    "RandomForest": probs_rf_test,
    "XGBoost": probs_xgb_test,
    "LightGBM": probs_lgb_test,
    "ANN": probs_ann_test,
    "DeepMLP": probs_dmlp_test,
    "TabTransformer": probs_tab_test,
    "Ensemble": probs_ens_test
}

# Combined ROC & PR
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
for name, probs in models_probs.items():
    fpr, tpr, _ = roc_curve(y_test_raw, probs)
    plt.plot(fpr, tpr, label=f"{name} (AUC={roc_auc_score(y_test_raw, probs):.3f})")
plt.plot([0,1],[0,1],'k--'); plt.title("Combined ROC"); plt.xlabel("FPR"); plt.ylabel("TPR"); plt.legend()
plt.subplot(1,2,2)
for name, probs in models_probs.items():
    prec, rec, _ = precision_recall_curve(y_test_raw, probs)
    plt.plot(rec, prec, label=f"{name} (AP={average_precision_score(y_test_raw, probs):.3f})")
plt.title("Combined Precision-Recall"); plt.xlabel("Recall"); plt.ylabel("Precision"); plt.legend()
plt.tight_layout(); plt.show()

# Combined calibration
plt.figure(figsize=(7,6))
for name, probs in models_probs.items():
    # frac, mean = calibration_curve(y_test_raw, probs, n_bins=10) # Removed due to import error
    # plt.plot(mean, frac, marker='o', label=name) # Removed due to import error
    pass # Keep ideal line handled in plot_dashboard for each model
plt.plot([0,1],[0,1],'k--'); plt.title("Calibration curves (test) - Plotting Skipped"); plt.xlabel("Mean pred"); plt.ylabel("Frac pos"); # plt.legend() # Removed as no models are plotted
plt.show()

# Results table (summary)
rows = []
for name, probs in models_probs.items():
    ta = threshold_analysis(probs, y_test_raw)
    rows.append({
        "model": name,
        "auc": float(roc_auc_score(y_test_raw, probs)),
        "ap": float(average_precision_score(y_test_raw, probs)),
        "acc@0.5": ta[0.5]["acc"] if 0.5 in ta else None,
        "f1@0.5": ta[0.5]["f1"] if 0.5 in ta else None,
        "best_f1_thresh": ta["best_f1_thresh"],
        "best_f1": ta["best_f1"]
    })
df_results = pd.DataFrame(rows).sort_values(by="auc", ascending=False).reset_index(drop=True)
display(df_results)
df_results.to_csv(os.path.join(OUTDIR, "results_table.csv"), index=False)

# ------------------------------
# 6) SHAP explainability (XGBoost & LightGBM)
# ------------------------------
# We'll compute SHAP for a subsample to keep runtime reasonable
n_shap = min(2000, X_test.shape[0])
X_test_df = pd.DataFrame(X_test, columns=FEAT_NAMES)
print("Running SHAP for XGBoost (subsample)...")
expl_xgb = shap.TreeExplainer(xgb_model)
shap_vals_xgb = expl_xgb.shap_values(X_test_df.iloc[:n_shap])
shap.summary_plot(shap_vals_xgb, X_test_df.iloc[:n_shap], feature_names=FEAT_NAMES, show=True)

print("Running SHAP for LightGBM (subsample)...")
expl_lgb = shap.TreeExplainer(lgb_model)
shap_vals_lgb = expl_lgb.shap_values(X_test_df.iloc[:n_shap])
shap.summary_plot(shap_vals_lgb, X_test_df.iloc[:n_shap], feature_names=FEAT_NAMES, show=True)

# Top features by mean abs SHAP
mean_abs_xgb = np.abs(shap_vals_xgb).mean(axis=0)
feat_rank_xgb = sorted(zip(FEAT_NAMES, mean_abs_xgb), key=lambda x: -x[1])[:15]
print("Top XGBoost SHAP features:", feat_rank_xgb)

mean_abs_lgb = np.abs(shap_vals_lgb).mean(axis=0)
feat_rank_lgb = sorted(zip(FEAT_NAMES, mean_abs_lgb), key=lambda x: -x[1])[:15]
print("Top LightGBM SHAP features:", feat_rank_lgb)

# ------------------------------
# 7) Ensemble ablation
# ------------------------------
base_models = {"XGBoost": xgb_model, "RandomForest": rf, "LightGBM": lgb_model}
abl_results = {}
for drop in base_models.keys():
    keep = {k: v for k, v in base_models.items() if k != drop}
    estimators = [(k, v) for k, v in keep.items()]
    ens_sub = VotingClassifier(estimators=estimators, voting="soft", n_jobs=-1)
    ens_sub.fit(X_train_res, y_train_res)
    probs = predict_proba_safe(ens_sub, X_test)
    abl_results[f"Ensemble_minus_{drop}"] = {"AUC": float(roc_auc_score(y_test_raw, probs)),
                                            "F1": float(f1_score(y_test_raw, (probs>=0.5).astype(int)))}
# full ensemble row
abl_results["Full_Ensemble"] = {"AUC": float(roc_auc_score(y_test_raw, probs_ens_test)),
                                "F1": float(f1_score(y_test_raw, (probs_ens_test>=0.5).astype(int)))}
df_abl = pd.DataFrame(abl_results).T
display(df_abl)
df_abl.to_csv(os.path.join(OUTDIR, "ensemble_ablation.csv"))

# ------------------------------
# 8) Calibration metrics (Brier & ECE) + DeLong AUC test
# ------------------------------
calib_rows = []
for name, probs in models_probs.items():
    brier = float(brier_score_loss(y_test_raw, probs))
    ece = float(expected_calibration_error(y_test_raw, probs, n_bins=10))
    calib_rows.append({"model": name, "brier": brier, "ece": ece})
df_calib = pd.DataFrame(calib_rows).sort_values("brier")
display(df_calib)
df_calib.to_csv(os.path.join(OUTDIR, "calibration_metrics.csv"), index=False)

# if _DELONG_AVAILABLE: # Removed due to import error
#     print("Running DeLong test (ensemble vs others)...")
#     for comp in ["XGBoost", "RandomForest", "LightGBM", "DeepMLP"]:
#         try:
#             pval = delong_roc_test(y_test_raw, probs_ens_test, models_probs[comp])
#             print(f"DeLong p-value (Ensemble vs {comp}): {pval}")
#         except Exception as e:
#             print("DeLong failed for", comp, str(e))
# else: # Keep this print statement
print("delong not available â€” skip DeLong tests. Install 'delong' pip package to enable.")


# ------------------------------
# 9) Threshold tuning & subgroup fairness (example: Sex, Age, Education)
# ------------------------------
print("Threshold tuning for Ensemble (grid 0.1..0.9)")
grid = np.linspace(0.1, 0.9, 9)
ta_grid = threshold_analysis(probs_ens_test, y_test_raw, ths=list(grid))
# Print best thresholds by metrics
best_f1_thr = ta_grid["best_f1_thresh"]
print("Ensemble best-F1 threshold:", best_f1_thr, "best-F1:", ta_grid["best_f1"])
# find threshold that gives recall >= 0.8 if exists
rec_table = {t: ta_grid[t]["rec"] for t in grid}
for t in sorted(rec_table, key=lambda x: -rec_table[x]):
    pass  # user can inspect rec_table or you can print top ones

# Subgroup fairness
subgroups = {"Sex": X_test_raw["Sex"], "Age": X_test_raw["Age"], "Education": X_test_raw["Education"]}
fairness_rows = []
for sg, vals in subgroups.items():
    for v in np.unique(vals):
        idx = (vals == v)
        if idx.sum() < 30:  # skip tiny groups
            continue
        aucv = float(roc_auc_score(y_test_raw[idx], probs_ens_test[idx]))
        f1v = float(f1_score(y_test_raw[idx], (probs_ens_test[idx] >= 0.5).astype(int)))
        fairness_rows.append({"subgroup": sg, "value": v, "n": int(idx.sum()), "auc": aucv, "f1": f1v})
df_fairness = pd.DataFrame(fairness_rows)
display(df_fairness)
df_fairness.to_csv(os.path.join(OUTDIR, "subgroup_fairness.csv"), index=False)

# ------------------------------
# 10) PDP & ALE (PDP via sklearn, ALE via PyALE if available)
# ------------------------------
# Choose top features (from SHAP) for PDP/ALE
top_xgb = [f for f,_ in feat_rank_xgb[:3]] if 'feat_rank_xgb' in globals() else FEAT_NAMES[:3]
feature_indices = [FEAT_NAMES.index(f) for f in top_xgb]

print("Partial Dependence Plots for top features:", top_xgb)
try:
    display(PartialDependenceDisplay.from_estimator(xgb_model, X_test, features=feature_indices, feature_names=FEAT_NAMES))
    plt.show()
except Exception as e:
    print("PDP failed:", e)

if _PYALE_AVAILABLE:
    print("Running ALE for top features (this may be slower)...")
    X_df_for_ale = pd.DataFrame(X_test, columns=FEAT_NAMES)
    for feat in top_xgb:
        try:
            print("ALE for", feat)
            ale(X=X_df_for_ale, model=xgb_model, feature=[feat])
        except Exception as e:
            print("ALE failed for", feat, e)
else:
    print("PyALE not installed: skip ALE. To enable, pip install PyALE.")

# ------------------------------
# 11) Stacking Ensemble & Surrogate rules
# ------------------------------
stack = StackingClassifier(
    estimators=[("xgb", xgb_model), ("rf", rf), ("lgb", lgb_model)],
    final_estimator=LogisticRegression(max_iter=1000),
    cv=5, n_jobs=-1
)
stack.fit(X_train_res, y_train_res)
probs_stack_test = predict_proba_safe(stack, X_test)
print("Stacking AUC (test):", roc_auc_score(y_test_raw, probs_stack_test))

# Surrogate decision tree for interpretability (learn ensemble decisions)
surrogate = DecisionTreeClassifier(max_depth=4, random_state=SEED)
surrogate.fit(X_test, (probs_ens_test >= 0.5).astype(int))
rules = export_text(surrogate, feature_names=FEAT_NAMES)
print("Surrogate tree rules:\n", rules)
with open(os.path.join(OUTDIR, "surrogate_rules.txt"), "w") as fh:
    fh.write(rules)

# ------------------------------
# 12) Save models & final combined CSV
# ------------------------------
joblib.dump(ensemble, os.path.join(OUTDIR, "ensemble.pkl"))
joblib.dump(stack, os.path.join(OUTDIR, "stacking.pkl"))
joblib.dump(xgb_model, os.path.join(OUTDIR, "xgb_model.pkl"))
joblib.dump(lgb_model, os.path.join(OUTDIR, "lgb_model.pkl"))
joblib.dump(rf, os.path.join(OUTDIR, "rf_model.pkl"))
# Save final combined results (merge df_results, df_calib, df_abl, feat_rank lists)
final = df_results.copy()
# add brier & ece by mapping
final = final.merge(df_calib, left_on="model", right_on="model", how="left")
# add top shap features for XGB/LGB
final["top_shap"] = None
if 'feat_rank_xgb' in globals():
    final.loc[final.model == "XGBoost", "top_shap"] = ", ".join([f"{f}:{round(v,3)}" for f,v in feat_rank_xgb[:10]])
if 'feat_rank_lgb' in globals():
    final.loc[final.model == "LightGBM", "top_shap"] = ", ".join([f"{f}:{round(v,3)}" for f,v in feat_rank_lgb[:10]])
final.to_csv(os.path.join(OUTDIR, "final_results_combined.csv"), index=False)
print("Saved final_results_combined.csv and models to", OUTDIR)

# ------------------------------
# 13) Deployment snippet (Streamlit) - printed for copy/paste
# ------------------------------
print("\n--- Streamlit deployment snippet (paste into app.py) ---\n")
print(r'''
import streamlit as st
import joblib
import pandas as pd

model = joblib.load("final_ensemble.pkl")  # or stacking.pkl
st.title("Heart Disease Risk Prediction (Demo)")
# collect inputs (example)
bmi = st.number_input("BMI", value=25.0)
age = st.number_input("Age", value=45)
# ... add inputs for all FEAT_NAMES in the correct order
if st.button("Predict"):
    # create dataframe with exactly the FEAT_NAMES ordering used in preprocessing
    df = pd.DataFrame([[...]], columns=[...])
    prob = model.predict_proba(df)[:,1][0]
    st.write("Risk probability:", round(prob, 4))
''')
```
